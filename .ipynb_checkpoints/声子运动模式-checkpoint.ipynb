# 轴向振动模式

## 求解平衡位置

$u_{j}-\sum_{n=1}^{j-1}1/(u_{j}-u_{n})^{2}+\sum_{n=j+1}^{N}1/(u_{j}-u_{n})^{2}=0,$
$u_{n}\equiv z_{n}^{0}/\ell$ $(n=1,2,\dots,N),$
where $\ell\equiv\sqrt[3]{e^{2}/4\pi\epsilon_{0}M\omega_{z}^{2}}$

```python
def position(x):
    # list 长度为离子数目, x list 是每个离子的位置初值，通过迭代法求解 x0 。
    n = len(x)
    e = np.zeros(n) #n个平衡方程
    for i in range(1,n-1):
        for j in range(0,i):
            e[i] = e[i] - 1/(x[i]-x[j])**2
        for k in range(i+1,n):
            e[i] = e[i] + 1/(x[i]-x[k])**2
        e[i] = e[i] + x[i]
    for i in range(1,n):
        e[0] = e[0] + 1/(x[0]-x[i])**2
    e[0] = e[0] + x[0]
    for i in range(0,n-1):
        e[n-1] = e[n-1] - 1/(x[n-1]-x[i])**2
    e[n-1] = e[n-1] + x[n-1]    
    return(np.array(e))

x0 = np.arange(4)  # 数值运算的初始值。 list 长度为离子数目
result = fsolve(position, x0)
print(result)
```

## 频率分布



x,y,z 方向的径向分布


$\begin{equation}
A_{nj}^{\xi}=\left\{ \begin{array}{ll}
\beta_{\xi}^{2}+\sum\limits _{p=1,p\not=j}^{N}a_{\xi}/|u_{j}-u_{p}|^{3} & (n=j)\\
-a_{\xi}/|u_{j}-u_{n}|^{3} & (n\not=j),
\end{array}\right.
\end{equation}$
with $\beta_{\xi}=\omega_{\xi}/\omega_{z}$, $a_{x}=a_{y}=-1$, and
$a_{z}=2$ 

```python
# 求解径向频率分布的改变, 注意 sqrt(f2) 是频率分布

def mode_x(n,b,DipoleList=None):
    #  n 为离子数
    #  b为trap参数， w_x/w_z的数值
    # DipoleList，形式为[(i,b_i),(j,b_j),....] 分别在 i个离子上施加b_i 的局域势...
    # DipoleList 为None时， 为普通阱
    A = np.zeros((n,n)) #展开矩阵
    u = fsolve(position, np.arange(n))
    for i in range(0,n):
        A[i][i] =  b**2
        for j in range(0,n):
            if i != j:
                A[i][j]  =  1/abs(u[i]-u[j])**3
                A[i][i] += -1/abs(u[i]-u[j])**3
    if DipoleList is not None:
        for k, b1 in DipoleList:
            A[k][k] = A[k][k] + b1**2 
    a,b = np.linalg.eig(A)
    return a,b

# 求轴向的频率分布
def mode_z(n,DipoleList=None):
    #  n 为离子数
    #  b为trap参数， w_x/w_z的数值
    # DipoleList，形式为[(i,b_i),(j,b_j),....] 分别在 i个离子上施加b_i 的局域势...
    # DipoleList 为None时， 为普通阱
    A = np.zeros((n,n)) #展开矩阵
    u = fsolve(position, np.arange(n))
    for i in range(0,n):
        A[i][i] = 1      # 对角项初始值， 便于后续积分
        for j in range(0,n):
            if i != j:
                A[i][j]  =   -2/abs(u[i]-u[j])**3  # 非对角项
                A[i][i] +=  2/abs(u[i]-u[j])**3  # 对角项需要积分， p=j
    
    if DipoleList is not None:
        for k, b1 in DipoleList:
            A[k][k] = A[k][k] + b1**2 
    a,b = np.linalg.eig(A)
    return a,b



#注意 sqrt(a) 是频率分布

def resortEig(M, indexM):
    # resort matrix eigs with index matrix or index vectors indexM[:,i]
    m,n = M.shape
    index = []
    for i in range(n):
        p = [np.abs(np.dot(M[:,j],indexM[:,i]))/np.linalg.norm(M[:,j])/np.linalg.norm(indexM[:,i]) for j in range(n)]
        index.append(np.argmax(p))
    return np.array(index)


def arangeEig(a,V):
    index = np.argsort(a)
    return (a[index], V[:,index])
```

## Diople Tailor Modes 

### Transverse Modes

```python
# assume the axial mode frequency is 1MHz

N = 21    # ion number
beta = 10.0


D1 = 9     # first ion dipole on
D2 = 8     # second ion dipole on
dipole_strength = 10.0

v_dipole_list = np.linspace(0,dipole_strength,100) # dipole strenth

a,indexM = mode_x(N,beta)
a,indexM = arangeEig(a, indexM)

v_list = []
for v in v_dipole_list:
    DipoleListX = [(D1,v),(D2,v)]
    f2, M = mode_x(n=N,b=beta,DipoleList=DipoleListX)
    index = resortEig(M,indexM)
    indexM = M[:,index]

    v_list.append(np.sqrt(f2)[index])
    
# # print(v_list)
datas = np.array(v_list)

for i in range(N):
    plt.plot(v_dipole_list, datas[:,N-i-1])  # 
    
    
# plt.plot(v_dipole_list,np.sqrt(beta**2+v_dipole_list**2-8))


plt.ylabel('motion frequency ($v_z$)')
plt.xlabel('dipole strength ($v_z$)')
plt.title('%d ions, dipoles on the %dth and  %dth ion, beta=%.1f'%(N, D1+1,D2+1,beta))

plt.savefig('%d ions  %d %d beta=%.1f.pdf'%(N,D1+1,D2+1,beta))
```

```python
N = 11    # ion number
beta = 10.0
v_dipole_list = np.linspace(0,10,100) # dipole strenth
D1 = 6      # the fisrt ion dipole1 on

def tailorGap(i,j, v_dipole_list):
    # i     # first ion dipole on
    # j     # second ion dipole on
    a,indexM = mode_x(N,beta)
    a,indexM = arangeEig(a, indexM)

    v_list = []
    for v in v_dipole_list:
        DipoleListX = [(i,v),(j,v)]
        f2, M = mode_x(n=N,b=beta,DipoleList=DipoleListX)
        index = resortEig(M,indexM)
        indexM = M[:,index]
        v_list.append(np.sqrt(f2)[index])
    datas = np.array(v_list)
    return datas[:,N-1] - datas[:,N-2]

for j in range(0,N):
    if j !=D1:
        plt.plot(v_dipole_list, tailorGap(D1,j, v_dipole_list), label='i=%d'%(j+1))


params = {
    'legend.fontsize': 5*2.54,
}
mpl.rcParams.update(params)    
plt.legend()
plt.title("N=%d, b=%.1f, first dipole on the %d'th ion, second dipole on i'th ion"%(N,beta,D1+1))
plt.ylabel('frequency splitting ($v_z$)')
plt.xlabel('dipole strength ($v_z$)')

plt.savefig("Gap: N=%d, b=%.1f, first dipoles on the %d'th ion, second dipole on i'th ion.png"%(N,beta,D1+1))
```

## 510nm dipole trap

```python
from sympy import *
init_printing()
from scipy.constants import c as c0
from scipy.constants import m_p
from scipy.constants import hbar as hbar_const

m,c,omega0,Gamma,Delta1,Delta2, I, a,  = symbols('M, c, \omega_0, \Gamma,\Delta_1, \Delta_2 I, a', real=True,positive=True)
wavelen1,wavelen2, wavelen_diploe, hbar, P = symbols('\lambda_1,\lambda_2,\lambda_d, hbar, P', real=True,positive=True)

U_dip = pi*c**2*Gamma/(2*omega0**3)*(2/Delta2+1/Delta1)*I

paras={
    Gamma: 20e6*2*pi,
    wavelen1: 0.369e-6,
    wavelen2: 0.329e-6,
    wavelen_diploe: 0.532e-6,
    omega0: c/wavelen1*2*pi,  # note! omega0
    Delta: 2*pi*200e9,
    Delta1: c0/wavelen1*2*pi-c0/wavelen_diploe*2*pi,
    Delta2: c0/wavelen2*2*pi-c0/wavelen_diploe*2*pi,
    a:    wavelen_diploe/0.4/2,   # 0.4NA, radius
    m:     171*m_p,
    P:     2000e-3,  # 100 mW， 554nm laser
    c:  c0,
    hbar: hbar_const,
}

I_0 = 2*P/(pi*a**2) 

U_0 = U_dip.subs({I:I_0})

w_dipole = sqrt(4*U_0/(m*a**2))

N(w_dipole.subs(paras).subs(paras)/2/pi/1e6)    #   z direction confined be light focus

# w_dipole_z
# w_dipole_z
```

```python
v_z = 2*np.pi*200e3    # v_z = 200Kz

DipoleList = [(9,20),(11,20)]
n = 21
b=10.0

C_eta = 462.391277099127  # eta = C_eta/sqrt(omega), get from 171Yb Raman excitaion @ 369nm, 

n1 = DipoleList[0][0]
n2 = DipoleList[1][0]
l,v = mode_x(n,b,DipoleList)
l,v = arangeEig(l, v)
# l = l[::-1]
# v = v[:,::-1]

v_list = np.sqrt(l) * v_z
eta_list = C_eta/np.sqrt(v_list)

g_list = eta_list*v


from scipy import integrate, interpolate

def get_alpha_final(j,k,A_t, dt):    # A_t  = A0(t) * sin(w*t+phi(t))
    # j th ion, k th mode 
    t = np.arange(len(A_t))*dt
    a = g_list[j,k]*A_t*np.exp(1.0j*v_list[k]*t)
    return np.sum(a)*dt

def get_alpha_list(j,k,A_t, dt):
     # j th ion, k th mode 
    t = np.arange(len(A_t))*dt
    a = g_list[j,k]*A_t*np.exp(1.0j*v_list[k]*t)
    return np.cumsum(a)*dt



def GetPhi_jn(j,n, k, A_j, A_n, t_list):
    
    dt = t_list[1]-t_list[0]
    
    Aj, An = np.meshgrid(A_j,A_n)   # j---t2, n---t1
    t2, t1  = np.meshgrid(t_list,t_list)

    M = 2*Aj*An*g_list[j,k]*g_list[n,k]*np.sin(v_list[k]*(t2-t1))
    return np.sum(np.triu(M))*dt*dt


# A_j = interpolate.interp1d(t,A_t)
# A_n = interpolate.interp1d(t,A_t)
# f = lambda t2, t1: 2*A_j(t2)*A_n(t1)*g_list[j,k]*g_list[n,k]*np.sin(v_list[k]*(t2-t1))
# integrate.dblquad(f, 0, t_f, lambda x: 0, lambda x: x)
```

## 优化函数

```python
%%time

v_z = 2*np.pi*200e3    # v_z = 200Kz
n_ions = 21
b=10.0

Omega_SDF =  846e3 *2 *np.pi # SDF induced Rabi 

dipole_strenth = 20
ion_i = 9
ion_j = 10

t_f = 10e-6
N_samples = 1001
# Delta = 200e9  # Raman 200GHz detuning
# Yb_171_spin_splitting = 12.642e9 



v_x = b*v_z
v_diploe = dipole_strenth*v_z


DipoleList = [(ion_i, dipole_strenth),(ion_j,dipole_strenth)]

l,v = mode_x(n_ions,b,DipoleList)
l,v = arangeEig(l, v)
l = l[::-1]
v = v[:,::-1]

v_list = np.sqrt(l) * v_z
eta_list = C_eta/np.sqrt(v_list)
g_list = eta_list*v


t_list = np.linspace(0,t_f,N_samples)
dt = t_list[1]-t_list[0]


def set_phi_list(phi_zips,t_list):
    phis = np.zeros(len(t_list))
    i0 = 0
    for t, phi in phi_zips:
        i_end = np.sum(t_list<=t)
        phis[i0:i_end-1] = phi
        i0=i_end-1
    return phis



def cal(v_drive,t1,phi1,t2,phi2):
    
#     print(v_drive,t1,phi1,t2,phi2)

    phi_zips = [(t1,0),(t2,phi1),(t_f-t2,phi2),(t_f-t1,phi1),(t_f,0)]
    phi_list = set_phi_list(phi_zips, t_list)
#     plt.plot(phi_list)

    A_t_real  = Omega_SDF*np.sin(t_list*v_drive + phi_list)   
    # A_t  = Omega_SDF*np.exp(-1.0j*(t_list*v_drive + phi_list))/2.0  # rotating wave appromix
    A_t = A_t_real

    alpha_M = [[get_alpha_final(j,k,A_t,dt) for k in range(n_ions)] for j in range(n_ions)]
    dphi = GetPhi_jn(ion_i,ion_j,1, A_t_real,A_t_real,t_list)-GetPhi_jn(ion_i,ion_j,0, A_t_real,A_t_real,t_list) # the first two modes is the tailored modes
    phi_err = np.abs(np.abs(dphi) - np.pi/4)
    apha_err = np.sum(np.abs(alpha_M))
    print(phi_err+apha_err)
    return phi_err+apha_err


v_0 = (v_list[0]+v_list[1])/2 + 2*np.pi/t_f
t1_0 = t_f*0.2
t2_0 = t_f*0.4
phi1_0 = 0.0
phi2_0 = 0.0

def cal2(x):
    v_drive,t1,phi1,t2,phi2 = x
    return cal(v_drive,t1,phi1,t2,phi2)

from scipy.optimize import minimize



result = minimize(cal2,[v_0,t1_0,phi1_0,t2_0,phi2_0],tol=1e-10)

result

# cal(v_0,t1_0,phi1_0,t2_0,phi2_0)
```
