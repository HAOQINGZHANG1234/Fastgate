```python
def boltz(x, a1, a2, x0, dx):
    return (a1-a2)/(1+np.exp((x-x0)/dx))+a2
def allo(x, a, b):
    return a*x**b
a1 = -0.02093
a2 = 0.99669
x0 = 2.92637
dx = 0.4648
a = 4.28223e-8
b = 2.62506
```

```python
import pickle
ion_number = range(1,28)
number = np.linspace(1.2,28,100)
F1 = data_frame['in_with']
F2 = data_frame['in_no']
f = data_frame['F']
ff = np.polyfit(ion_number, f, 3)
pf = np.poly1d(f1)
yvalsf = pf(ion_number) 
plt.figure(figsize=(12,8))
fig, ax1 = plt.subplots()
ax2=ax1.twinx()
ax1.plot(ion_number, F1, 'o',ms=5,markeredgewidth=2, markeredgecolor='#3F51B5', markerfacecolor='#BBC2E7')
ax1.plot(ion_number, F2, 'o',ms=5,markeredgewidth=2, markeredgecolor='#80ce72', markerfacecolor='#d5efd1')
ax1.plot(number, boltz(number, a1, a2, x0, dx), color ="#80ce72",lw=2.5)
ax1.plot(number, allo(number, a, b), color ="#3F51B5",lw=2.5)
ax1.set_yscale("log")
ax1.set_ylim(1e-10,2)
ax1.set_xlabel('ion distance')
ax1.set_ylabel('gate infidelity')
ax2.plot(ion_number, f, 'o',ms=5,markeredgewidth=2, markeredgecolor='#d198db', markerfacecolor='#f6e4ef')
# ax2.plot(ion_number, f, color='xkcd:amber',linestyle="--",lw=2.5)
ax2.plot(ion_number, yvalsf, color='#d198db', lw=2.5)
ax2.set_ylabel(r'$\frac{\Omega_{SDF}}{2\pi} (MHz)$')
pickle.dump(fig,open('Fig.matplot','wb'))
```
