```python
import numpy as np
from scipy.optimize import fsolve
%matplotlib inline
import matplotlib.pyplot as plt
def position(x):
    n = len(x)
    e = np.zeros(n) #n个平衡方程
    for i in range(1,n-1):
        for j in range(0,i):
            e[i] = e[i] - 1/(x[i]-x[j])**2
        for k in range(i+1,n):
            e[i] = e[i] + 1/(x[i]-x[k])**2
        e[i] = e[i] + x[i]
    for i in range(1,n):
        e[0] = e[0] + 1/(x[0]-x[i])**2
    e[0] = e[0] + x[0]
    for i in range(0,n-1):
        e[n-1] = e[n-1] - 1/(x[n-1]-x[i])**2
    e[n-1] = e[n-1] + x[n-1]    
    return(np.array(e))
result = fsolve(position, [0,1,2,3])
print(result)
```

## 径向运动模式

```python
# 求解径向频率分布
def Frequence_rr(n,m1,m2,b,b1,b2):  # b为w_r/w_z
#     A_z = np.zeros((n,n)) #展开矩阵
    A_r = np.zeros((n,n))
    u = fsolve(position, np.linspace(0,n-1,n))
    for i in range(0,n):
        for j in range(0,n):
            if i != j:
#                 A_z[i][j] = -2/abs(u[i]-u[j])**3
#                 A_z[i][i] = A_z[i][i] + 2/abs(u[i]-u[j])**3
                A_r[i][j] = 1/abs(u[i]-u[j])**3
                A_r[i][i] = A_r[i][i] - 1/abs(u[i]-u[j])**3
#         A_z[i][i] = A_z[i][i] + 1
        A_r[i][i] = A_r[i][i] + b**2
#     A_z[m1][m1] = A_z[m1][m1] + b1**2
    A_r[m1][m1] = A_r[m1][m1] + b1**2 
#     A_z[m2][m2] = A_z[m2][m2] + b3**2
    A_r[m2][m2] = A_r[m2][m2] + b2**2 
#     a_z,b_z = np.linalg.eig(A_z)
    a_r,b_r = np.linalg.eig(A_r)
    return a_r,b_r
np.sqrt(Frequence_rr(20,10,11,10,10,10)[0])
```

# 积分

```python
def f2(w,u,t1,t2,p): ## t1>t2
    x = 0
    x = x - (u + w) * np.sin(p + t1 * (u - w))
    x = x + (u + w) * np.sin(p + t2 * (u - w))
    x = x + (u - w) * np.sin(p + t1 * (u + w))
    x = x - (u - w) * np.sin(p + t2 * (u + w))
    return -1/(2 * (u - w)*(u + w)) * x
# p 方向displacement要用的积分
f2(1,1.7,2,1,1)
```

```python
def f1(w,u,t1,t2,p):
    x = 0
    x = x + (u + w) * np.cos(p + t1 * (u - w))
    x = x - (u + w) * np.cos(p + t2 * (u - w))
    x = x + (u - w) * np.cos(p + t1 * (u + w))
    x = x - (u - w) * np.cos(p + t2 * (u + w))
    return -1/(2 * (u - w) * (u + w)) * x
# x 方向displacement要用的积分
f1(2.5,2,2,1,0.5)
```

```python
def f3(w,u,t1,t2,p): #这种情形是积分两端时间都在同一个区间 开始时间为t2，终止时间为t1
    x = 0
    x = x + w * (u - w) * (u + w) * (np.sin(2*(p + t1*u)) - np.sin(2*(p + t2*u)))
    x = x + u * (u + w) **2 * np.sin((t1 - t2) * (u - w)) - u * (u - w) **2 * np.sin((t1 - t2) * (u + w))
    x = x - 2 * w * u * (u + w) * (u - w) * (t1 - t2)
    x = x - (np.sin(2*p + t2 * (u - w) + t1 * (u + w)) - np.sin(2*p + t1 * (u - w) + t2 * (u + w))) * u *(u - w) * (u + w) 
    return 1/(4 * u * (u - w)**2 * (u + w)**2) * x
# 积累出的相位要用的积分    
f3(3,1,2,1,2.5)
```

```python
def f4(w,u,t1,t2,t3,t4,p1,p2): #这种情形是积分两端时间不在同一个区间 第一个离子的区间在第二个离子区间后 开始时间为t2，终止时间为t1 开始时间为t4，终止时间为t3 
    x = 0
    x = x - (u + w)**2 * (np.sin(p1 - p2 + (t1 - t3) * (u - w)) - np.sin(p1 - p2 + (t2 - t3) * (u - w)))
    x = x + (u + w)**2 * (np.sin(p1 - p2 + (t1 - t4) * (u - w)) - np.sin(p1 - p2 + (t2 - t4) * (u - w)))
    x = x + (u - w)**2 * (np.sin(p1 - p2 + (t1 - t3) * (u + w)) - np.sin(p1 - p2 + (t2 - t3) * (u + w)))
    x = x - (u - w)**2 * (np.sin(p1 - p2 + (t1 - t4) * (u + w)) - np.sin(p1 - p2 + (t2 - t4) * (u + w)))
    x = x + (u**2 - w**2) * (np.sin(p1 + p2 + u * (t1 + t3) + w * (t1 - t3)))
    x = x - (u**2 - w**2) * (np.sin(p1 + p2 + u * (t2 + t3) + w * (t2 - t3)))
    x = x - (u**2 - w**2) * (np.sin(p1 + p2 + u * (t1 + t3) - w * (t1 - t3)))
    x = x + (u**2 - w**2) * (np.sin(p1 + p2 + u * (t2 + t3) - w * (t2 - t3)))
    x = x - (u**2 - w**2) * (np.sin(p1 + p2 + u * (t1 + t4) + w * (t1 - t4)))
    x = x + (u**2 - w**2) * (np.sin(p1 + p2 + u * (t2 + t4) + w * (t2 - t4)))
    x = x + (u**2 - w**2) * (np.sin(p1 + p2 + u * (t1 + t4) - w * (t1 - t4)))
    x = x - (u**2 - w**2) * (np.sin(p1 + p2 + u * (t2 + t4) - w * (t2 - t4)))
    return 1/(4 * (u - w)**2 * (u + w)**2) * x
# 积累出的相位要用的积分    
f4(3,1,2,1,0.3,0.1,2.5,1.5)
```

# coupling

```python
def Coupling(n,n1,n2,b,w1,w2): #n1,n2 为加上激光的离子，n为离子总数,返回耦合系数，解出的频率本征值，频率本征态
    g = np.zeros((2,n))
    b1 = w1/(2*np.pi*1e5)
    b2 = w2/(2*np.pi*1e5)
    F = Frequence_rr(n,n1,n2,b,b1,b2)
    w_r = np.sqrt(F[0])*2*np.pi*1e5 
    b_r = F[1].T
    for i in range(0,n): ##第i个模式与第n1,n2个离子耦合
        g[0][i] = c_lamb * b_r[i][n1] / np.sqrt(w_r[i])
        g[1][i] = c_lamb * b_r[i][n2] / np.sqrt(w_r[i])
    return g,w_r,b_r
Coupling(100,49,50,15,0*2*np.pi*1.5*1e6,0*2*np.pi*1.5*1e6)[0]
```

# displacement

```python
from math import pow
def Displace(g,w_r,t_list,phi_list,u): # tao为门时间，u为激光之间失谐, 返回失谐模平方
    n = len(w_r)
    m = len(phi_list)
    a = np.zeros((2,n))
    b = np.zeros(n)
    T_0 = 0
    T_1 = 0
    T_p = 0
    T_m = 0
    for i in range(0,n):
        jf1 = 0
        jf2 = 0
        mu = pow(3, w_r[i]/ (4*np.pi*1e5))
        b[i] = (mu + 1/mu)/ (mu - 1/mu)/ 2
        for j in range(0,m): 
            jf1 = jf1 + f1(w_r[i],u,t_list[j+1],t_list[j],phi_list[j])
            jf2 = jf2 + f2(w_r[i],u,t_list[j+1],t_list[j],phi_list[j])
        a[0][i] = g[0][i]**2 * (jf1**2 + jf2**2)
        a[1][i] = g[1][i]**2 * (jf1**2 + jf2**2)
        T_0 = T_0 - a[0][i] * b[i]
        T_1 = T_1 - a[1][i] * b[i]
        T_p = T_p - (g[0][i]+g[1][i])**2 * (jf1**2 + jf2**2) * b[i]
        T_m = T_m - (g[0][i]-g[1][i])**2 * (jf1**2 + jf2**2) * b[i]
    
    return a, T_0, T_1, T_p, T_m
```

# Phi

```python
def Phi(g,w_r,t_list,phi_list,u):
    n = len(w_r)
    m = len(phi_list)
    phi = 0
    
    for i in range(0,n):
        jf = 0
        for j in range(0,m):
          
            jf  = jf + f3(w_r[i],u,t_list[j+1],t_list[j],phi_list[j])
            for k in range(0,j):
                jf = jf + f4(w_r[i],u,t_list[j+1],t_list[j],t_list[k+1],t_list[k],phi_list[j],phi_list[k])
        phi = phi + 2*g[0][i]*g[1][i]*jf
    return phi
g = np.ones((10,10))
w_r = np.ones(10)
t_list = [2,2.9,4]
phi_list = [1,1]
u =2
Phi(g,w_r,t_list,phi_list,u)
```

```python
def Coupling_high(n,n1,n2,b,w1,w2,a):
    g = np.zeros((2,n))
    b1 = w1/(2*np.pi*1e5)*np.sqrt((1-1/2*a**2))
    b2 = w2/(2*np.pi*1e5)*np.sqrt((1-1/2*a**2))
    F = Frequence_rr(n,n1,n2,b,b1,b2)
    w_r = np.sqrt(F[0])*2*np.pi*1e5 
    b_r = F[1].T
    for i in range(0,n): ##第i个模式与第n1,n2个离子耦合
        g[0][i] = c_lamb * b_r[i][n1] / np.sqrt(w_r[i])
        g[1][i] = c_lamb * b_r[i][n2] / np.sqrt(w_r[i])
    return g,w_r,b_r

def Coupling_exp(n,n1,n2,b,w1,w2,a):
    g = np.zeros((2,n))
    b1 = w1/(2*np.pi*1e5)*np.sqrt((1-np.exp(-(a)**2))/a**2)
    b2 = w2/(2*np.pi*1e5)*np.sqrt((1-np.exp(-(a)**2))/a**2)
    F = Frequence_rr(n,n1,n2,b,b1,b2)
    w_r = np.sqrt(F[0])*2*np.pi*1e5 
    b_r = F[1].T
    for i in range(0,n): ##第i个模式与第n1,n2个离子耦合
        g[0][i] = c_lamb * b_r[i][n1] / np.sqrt(w_r[i])
        g[1][i] = c_lamb * b_r[i][n2] / np.sqrt(w_r[i])
    return g,w_r,b_r

def Coupling_tweezer(n,n1,n2,b,w1,w2,u,a):
    u = u * 2 * np.pi * 1e6
    g = np.zeros((2,n))
    b1 = w1/(2*np.pi*1e5)
    b2 = w2/(2*np.pi*1e5)
    F = Frequence_rr(n,n1,n2,b,b1,b2)
    w_r = np.sqrt(F[0])*2*np.pi*1e5 
    b_r = F[1].T
    for i in range(0,n): ##第i个模式与第n1,n2个离子耦合
        g[0][i] = c_lamb * b_r[i][n1] / np.sqrt(w_r[i])
        g[1][i] = c_lamb * b_r[i][n2] / np.sqrt(w_r[i])
    d_0 = w_r[0]-u
    d_1 = w_r[1]-u
    t_0 = np.power(abs((d_0-a*u)/(d_0+a*u)),1/4)
    t_1 = np.power(abs((d_1-a*u)/(d_1+a*u)), 1/4)
    w_r[0] = u + np.sqrt(abs(d_0**2 - (a*u)**2))
    w_r[1] = u + np.sqrt(abs(d_1**2 - (a*u)**2))
    g[0][0] = g[0][0] / t_0
    g[1][0] = g[1][0] / t_0
    g[0][1] = g[0][1] / t_1
    g[1][1] = g[1][1] / t_1
    return g,w_r,b_r
```

## 计算高阶修正的函数

```python
def high_order_fie(C,para,tao,Np,u):
    u = u * 2 * np.pi * 1e6
    t_list = (np.array([0,para[0],tao/2,tao-para[0],tao])*1e-6).tolist()
    phi_list = [para[1],para[2],-para[2],-para[1]]
    g = C[0]
    w_r = C[1]
    Omega = np.sqrt(np.pi/(4*abs(Phi(C[0],C[1],t_list,phi_list,u))))
    a = np.zeros((4,n,1000))
    x0 = np.zeros(n)
    dis = []
    k0 = 0
    flag = 0
    T = np.linspace(0,tao-0.00001,1000)
    jf = 0 
    for k,t in enumerate(T):
        pos = 0
        while (t*1e-6 >= t_list[pos]):
            pos = pos+1
            if pos == 2 and flag == 0:
                k0 = k
                flag = 1
        for i in range(0,n):
            x0[i] = C_lamb/np.sqrt(w_r[i])
            jf_x = 0
            jf_y = 0
            jf = 0
            for j in range(0,pos-1): 
                jf_x = jf_x + f1(w_r[i],u,t_list[j+1],t_list[j],phi_list[j])
                jf_y = jf_y + f2(w_r[i],u,t_list[j+1],t_list[j],phi_list[j])
                jf = jf + f3(w_r[i],u,t_list[j+1],t_list[j],phi_list[j])
                for l in range(0,j):
                    jf = jf + f4(w_r[i],u,t_list[j+1],t_list[j],t_list[l+1],t_list[l],phi_list[j],phi_list[l])
            jf_x = jf_x + f1(w_r[i],u,t*1e-6,t_list[pos-1],phi_list[pos-1])
            jf_y = jf_y + f2(w_r[i],u,t*1e-6,t_list[pos-1],phi_list[pos-1])
            jf = jf + f3(w_r[i],u,t*1e-6,t_list[pos-1],phi_list[pos-1])
            for l in range(0,pos-1):
                jf = jf + f4(w_r[i],u,t*1e-6,t_list[pos-1],t_list[l+1],t_list[l],phi_list[pos-1],phi_list[l])
            a[0][i][k] = g[0][i] * jf_x * Omega * x0[i] # 第一个离子第i个模式x方向位移
            a[1][i][k] = g[0][i] * jf_y * Omega * x0[i] # 第一个离子第i个模式y方向位移
            dis.append(a[0][i][k])
            dis.append(a[1][i][k])
```

```python
def FIE(n,n1,n2,b,w1,w2,para,tao,Np,u): #计算严格求解的非保真度，在考虑局域势的效益下，para是所有参数，tao是总时间，u是失谐，Np是门之前剩余热分布
    u = u * 2 * np.pi * 1e6
    w1 = w1 * 1e6 * 2 * np.pi
    w2 = w2 * 1e6 * 2 * np.pi
#     t_list = (np.array([0,para[0],tao/2,tao-para[0],tao])*1e-6).tolist()
#     phi_list = [para[1],para[2],-para[2],-para[1]]
#     t_list = (np.array([0,para[0],para[0]+para[1],tao/2,tao-para[0]-para[1],tao-para[0],tao])*1e-6).tolist() 
    t_list = (np.array([0,tao/6,tao/3,tao/2,2*tao/3,5*tao/6,tao])*1e-6).tolist()
    phi_list = [para[0],para[1],para[2],-para[2],-para[1],-para[0]]
    C = Coupling(n,n1,n2,b,w1,w2)
    b_r = C[2]
    Omega_pf = np.pi / ( 4 * abs(Phi(C[0],C[1],t_list,phi_list,u)) )
#     Omega_pf = (2*np.pi*2*1e6)**2
#     phi = Phi(C[0],C[1],t_list,phi_list,u)*Omega_pf
    D = (np.array(Displace(C[0],C[1],t_list,phi_list,u)) * Omega_pf).tolist()
    F = 1/6 * (2*(np.exp(D[1]) + np.exp(D[2])) + np.exp(D[3]) + np.exp(D[4]))  
    return F,D[0][1]+D[0][0],C[1]/(2*np.pi*1e6),np.sqrt(Omega_pf)/(2*np.pi*1e6)


def TRAJECT(n,n1,n2,b,w1,w2,para,tao,Np,u):
    u = u * 2 * np.pi * 1e6
    w1 = w1 * 1e6 * 2 * np.pi
    w2 = w2 * 1e6 * 2 * np.pi
#     t_list = (np.array([0,para[0],tao/2,tao-para[0],tao])*1e-6).tolist()
#     phi_list = [para[1],para[2],-para[2],-para[1]]
#     t_list = (np.array([0,para[0],para[0]+para[1],tao/2,tao-para[0]-para[1],tao-para[0],tao])*1e-6).tolist()
#     phi_list = [para[2],para[3],para[4],-para[4],-para[3],-para[2]]
    t_list = (np.array([0,tao/6,tao/3,tao/2,2*tao/3,5*tao/6,tao])*1e-6).tolist()
    phi_list = [para[0],para[1],para[2],-para[2],-para[1],-para[0]]
    C = Coupling(n,n1,n2,b,w1,w2)
    g = C[0]
    w_r = C[1]
    Omega = np.sqrt(np.pi / ( 4 * abs(Phi(C[0],C[1],t_list,phi_list,u)) ))
#     Omega = np.pi*2*2e6
    a = np.zeros((6,n,1000))
    b = np.zeros(1000)
    k0 = 0
    flag = 0
    k1 = 0
    flag_1 = 0
    T = np.linspace(0,tao-0.00001,1000)
    jf = 0 
    for k,t in enumerate(T):
        pos = 0
        while (t*1e-6 >= t_list[pos]):
            pos = pos+1
            if pos == 2 and flag == 0:
                k0 = k
                flag = 1
            if pos == 3 and flag_1 == 0:
                k1 = k
                flag_1 = 1
        for i in range(0,n):
            jf_x = 0
            jf_y = 0
            jf = 0
            for j in range(0,pos-1): 
                jf_x = jf_x + f1(w_r[i],u,t_list[j+1],t_list[j],phi_list[j])
                jf_y = jf_y + f2(w_r[i],u,t_list[j+1],t_list[j],phi_list[j])
                jf = jf + f3(w_r[i],u,t_list[j+1],t_list[j],phi_list[j])
                for l in range(0,j):
                    jf = jf + f4(w_r[i],u,t_list[j+1],t_list[j],t_list[l+1],t_list[l],phi_list[j],phi_list[l])
            jf_x = jf_x + f1(w_r[i],u,t*1e-6,t_list[pos-1],phi_list[pos-1])
            jf_y = jf_y + f2(w_r[i],u,t*1e-6,t_list[pos-1],phi_list[pos-1])
            jf = jf + f3(w_r[i],u,t*1e-6,t_list[pos-1],phi_list[pos-1])
            for l in range(0,pos-1):
                jf = jf + f4(w_r[i],u,t*1e-6,t_list[pos-1],t_list[l+1],t_list[l],phi_list[pos-1],phi_list[l])
            a[0][i][k] = g[0][i] * jf_x * Omega # 第一个离子第i个模式x方向位移
            a[1][i][k] = g[0][i] * jf_y * Omega # 第一个离子第i个模式y方向位移
            a[2][i][k] = g[1][i] * jf_x * Omega # 第二个离子第i个模式x方向位移
            a[3][i][k] = g[1][i] * jf_y * Omega # 第二个离子第i个模式y方向位移
            b[k] = b[k] + 2*g[0][i]*g[1][i]*Omega**2*jf

    fig = plt.figure(figsize=(10,10))
    plt.subplot(2,2,1)
    plt.plot(a[0][0][0:k0-1],a[1][0][0:k0-1],alpha=0.6,linewidth=2,c='blue')
    plt.plot(a[0][0][k0:k1-1],a[1][0][k0:k1-1],alpha=0.6,linewidth=2,c='green')
    plt.plot(a[0][0][k1:500],a[1][0][k1:500],alpha=0.6,linewidth=2,c='red')
    plt.plot(a[0][0][500:1000-k0-k1],a[1][0][500:1000-k0-k1],alpha=0.9,linewidth=2,c='red')
    plt.plot(a[0][0][1000-k0-k1:1000-k0],a[1][0][1000-k0-k1:1000-k0],alpha=0.9,linewidth=2,c='green')
    plt.plot(a[0][0][1000-k0:1000],a[1][0][1000-k0:1000],alpha=0.9,linewidth=2,c='blue')
    plt.plot(a[0][0][0],a[1][0][0],'o',markersize=10,c='skyblue')
    plt.plot(a[0][0][999],a[1][0][999],'*',markersize=10,c='magenta')
    plt.text(-0.85, -0.75, '$ \\alpha_j^k$', fontsize=20)
    plt.vlines(0,-0.5,0.5,linewidth=0.5)
    plt.hlines(0,-0.5,0.5,linewidth=0.5)
    plt.xlim(-0.5,0.5)
    plt.ylim(-0.5,0.5)
    
    plt.subplot(2,2,2)
    plt.plot(a[0][1][0:k0-1],a[1][1][0:k0-1],alpha=0.6,linewidth=2,c='blue')
    plt.plot(a[0][1][k0:k1-1],a[1][1][k0:k1-1],alpha=0.6,linewidth=2,c='green')
    plt.plot(a[0][1][k1:500],a[1][1][k1:500],alpha=0.6,linewidth=2,c='red')
    plt.plot(a[0][1][500:1000-k0-k1],a[1][1][500:1000-k0-k1],alpha=0.9,linewidth=2,c='red')
    plt.plot(a[0][1][1000-k0-k1:1000-k0],a[1][1][1000-k0-k1:1000-k0],alpha=0.9,linewidth=2,c='green')
    plt.plot(a[0][1][1000-k0:1000],a[1][1][1000-k0:1000],alpha=0.9,linewidth=2,c='blue')
    plt.plot(a[0][1][0],a[1][1][0],'o',markersize=10,c='skyblue')
    plt.plot(a[0][1][999],a[1][1][999],'*',markersize=10,c='magenta')
    plt.vlines(0,-0.5,0.5,linewidth=0.5)
    plt.hlines(0,-0.5,0.5,linewidth=0.5)
    plt.xlim(-0.5,0.5)
    plt.ylim(-0.5,0.5)
    
    ax = plt.subplot(2,2,3)
    ax.plot(a[0][0][0],a[0][0][0],'o',markersize=10)
    for i in range(2,10):
    #     plt.plot(pos1_x[i],pos1_y[i])
        ax.plot(a[0][i][999],a[1][i][999],'*',markersize=10)
        ax.ticklabel_format(axis='y',style='sci',scilimits=(0,3))
    ax.ticklabel_format(axis='x',style='sci',scilimits=(0,3))  
    plt.xlim(-0.001,0.001)
    plt.ylim(-0.001,0.001)
    
    fig.subplots_adjust(wspace=0.3, hspace=0.3)
    
    plt.subplot(2,2,4)
    plt.plot(T[0:k0-1],b[0:k0-1],alpha=0.6,linewidth=2)
    plt.plot(T[k0:k1-1],b[k0:k1-1],alpha=0.6,linewidth=2)
    plt.plot(T[k1:500],b[k1:500],alpha=0.6,linewidth=2)
    plt.plot(T[500:1000-k0-k1],b[500:1000-k0-k1],alpha=0.6,linewidth=2)
    plt.plot(T[1000-k0-k1:1000-k0],b[1000-k0-k1:1000-k0],alpha=0.6,linewidth=2)
    plt.plot(T[1000-k0:1000],b[1000-k0:1000],alpha=0.6,linewidth=2)
    plt.plot(T[0],b[0],'*',markersize=10,c='skyblue')
    plt.plot(T[999],b[999],'*',markersize=10,c='magenta')
    my_y_ticks = [0,-np.pi/4]
    plt.yticks(my_y_ticks,('$0$','$-\pi/4$'))
    
    
#     plt.xlim(-2.5,2.5)
#     plt.ylim(-2.5,2.5)
#     plt.subplot(2,2,2)
#     plt.plot(a[0][1][0:k0],a[1][1][0:k0])
#     plt.plot(a[0][1][k0:150],a[1][1][k0:150])
#     plt.plot(a[0][1][150:300-k0],a[1][1][150:300-k0])
#     plt.plot(a[0][1][300-k0:300],a[1][1][300-k0:300])
#     plt.plot(a[0][1][0],a[1][1][0],'*')
#     plt.plot(a[0][1][299],a[1][1][299],'*')
#     plt.xlim(-1.5,1.5)
#     plt.ylim(-1.5,1.5)
#     plt.subplot(2,2,3)
#     plt.plot(a[0][0]+a[2][0],a[1][0]+a[3][0])
#     plt.plot(a[0][0][0]+a[2][0][0],a[1][0][0]+a[3][0][0],'*')
#     plt.plot(a[0][0][299]+a[2][0][299],a[1][0][299]+a[3][0][299],'*')
#     plt.subplot(2,2,4)
#     plt.plot(a[0][1]+a[2][1],a[1][1]+a[3][1])
#     plt.plot(a[0][1][0]+a[2][1][0],a[1][1][0]+a[3][1][0],'*')
#     plt.plot(a[0][1][299]+a[2][1][299],a[1][1][299]+a[3][1][299],'*')
# #     plt.xlim(-1.5,1.5)
#     plt.ylim(-1.5,1.5)
#     plt.subplot(2,2,3)
#     plt.plot(a[2][0][0:k0],a[3][0][0:k0])
#     plt.plot(a[2][0][k0:150],a[3][0][k0:150])
#     plt.plot(a[2][0][150:300-k0],a[3][0][150:300-k0])
#     plt.plot(a[2][0][300-k0:300],a[3][0][300-k0:300])
#     plt.plot(a[2][0][0],a[3][0][0],'*')
#     plt.plot(a[2][0][299],a[3][0][299],'*')
#     plt.xlim(-1.5,1.5)
#     plt.ylim(-1.5,1.5)
#     plt.subplot(2,2,4)
#     plt.plot(a[2][1][0:k0],a[3][1][0:k0])
#     plt.plot(a[2][1][k0:150],a[3][1][k0:150])
#     plt.plot(a[2][1][150:300-k0],a[3][1][150:300-k0])
#     plt.plot(a[2][1][300-k0:300],a[3][1][300-k0:300])
#     plt.plot(a[2][1][0],a[3][1][0],'*')
#     plt.plot(a[2][1][299],a[3][1][299],'*')
#     plt.xlim(-1.5,1.5)
#     plt.ylim(-1.5,1.5)
#     plt.subplot(3,2,3)
#     plt.plot(a[2][0],a[3][0])
#     plt.plot(a[2][0][0],a[3][0][0],'*')
#     plt.plot(a[2][0][199],a[3][0][199],'*')
#     plt.subplot(3,2,4)
#     plt.plot(a[2][1],a[3][1])
#     plt.plot(a[2][1][0],a[3][1][0],'*')
#     plt.plot(a[2][1][199],a[3][1][199],'*')
#     plt.subplot(3,2,5)
#     plt.plot(a[4][1],a[3][1])
#     plt.plot(a[2][1][0],a[3][1][0],'*')
#     plt.plot(a[2][1][199],a[3][1][199],'*')
    
```
